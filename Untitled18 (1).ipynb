```python
import numpy as np
import matplotlib.pyplot as plt

def entropy(p):
    p = np.array(p)
    return -np.sum(p * np.log(p + 1e-12))  # avoid log(0)

def entropy_deficit(p):
    N = len(p)
    return np.log(N) - entropy(p)

def zipf_distribution(n, alpha=1.1):
    k = np.arange(1, n+1)
    p = 1 / k**alpha
    return p / p.sum()

def local_beta(delta_s, n_values):
    log_n = np.log(n_values)
    log_ds = np.log(delta_s)
    d_log_ds = np.diff(log_ds)
    d_log_n = np.diff(log_n)
    beta = -d_log_ds / d_log_n
    return np.append(beta, beta[-1])

def plot_cassini(beta_complex, b=1, s=0):
    x, y = np.meshgrid(np.linspace(-0.5, 1.5, 400), np.linspace(-1, 1, 400))
    z = x + 1j*y
    k2 = abs((beta_complex - b)*(beta_complex - s))
    f = abs((z - b)*(z - s))
    plt.contour(x, y, f, levels=[k2], colors='blue')
    plt.plot(np.real(beta_complex), np.imag(beta_complex), 'ro')
    plt.xlabel("Re(β)")
    plt.ylabel("Im(β)")
    plt.title("Cassini Oval: Entropy Fingerprint")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
n_values = np.arange(10, 201, 10)
distributions = [zipf_distribution(n) for n in n_values]
delta_s = [entropy_deficit(p) for p in distributions]
beta_vals = local_beta(delta_s, n_values)
beta_complex = np.mean(beta_vals) + 1j * np.std(beta_vals)
print("Entropy fingerprint:", beta_complex)
plot_cassini(beta_complex)
```
